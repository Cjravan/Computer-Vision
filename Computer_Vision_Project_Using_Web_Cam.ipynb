```python
# This is the code from which we can see all the color scales in one go

color_list=[method for method in dir(cv2) if method.startswith("COLOR_")is True]
color_list
len(color_list)

img=cv2.imread("Kiara-Advani-5.jpg",1)

for color in range(len(color_list)):
    col_name = color_list[color]
    col_code = "cv2." + col_name
    resized = cv2.resize(img,(400,400))
    try:
        rgb_image = cv2.cvtColor(resized, eval(col_code))
        cv2.imshow(col_name + "_"+ str(color), rgb_image)
        cv2.waitKey()
    except Exception as e:
        print("Error on", str(col_code))
        pass
    cv2.destroyAllWindows()
```

```python
# Important Tip:
# Wgenever you try to detect edges on your photos always change 
# it to black nd white by simply putting 0

import cv2
import numpy as pn

image = cv2.imread("kiara-advani-looks-phenomenal-as-ever-202212-1673076764.jpg",0)
cv2.imshow("Original Image",image)
cv2.waitKey()

height,width=image.shape

# extract sobel images
sobel_x=cv2.Sobel(image, cv2.CV_64F, 0,1,  ksize=5)
sobel_y=cv2.Sobel(image, cv2.CV_64F, 1,0,  ksize=5)

cv2.imshow("Sobel_X", sobel_x)
cv2.waitKey()

cv2.imshow("Sobel_Y", sobel_y)
cv2.waitKey()

sobel_OR=cv2.bitwise_or(sobel_x, sobel_y)
cv2.imshow("Sobel_x_y", sobel_OR)
cv2.waitKey()

laplacian= cv2.Laplacian(image, cv2.CV_64F)
cv2.imshow("laplacian", laplacian)
cv2.waitKey()

canny= cv2.Canny(image, 50,120)
cv2.imshow("Canny", canny)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# Mini Project - Live Sketch Using  WebCam
```

```python
# Image - GrayScale - Gausian Blur - Canny Edge extraction - threshold (inverse)

import cv2
import numpy as pn


# Our sketcg generation function
def sketch(image):
    # Part 1 - convert image to grayscale
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Part 2 - Clean up image using Gaussian Blur
    img_gray_blur = cv2.GaussianBlur(img_gray,(5,5),0)
    
    # Part 3 - Extract Edge - Canny
    Canny_edges = cv2.Canny(img_gray_blur,10,70)
    
    # Part 4 - Threshold (inverse)
    
    ret, mask =  cv2.threshold(Canny_edges, 70,255, cv2.THRESH_BINARY_INV)
    return mask

# Engaging Camera 
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv2.imshow("Mera Chitra", sketch(frame))
    if cv2.waitKey(1) == 13:
        break
        
# release camera and close windows
cap.release()

cv2.destroyAllWindows()
```
